```python
#nc = Dataset('grid_0.4min.nc','r')
import xarray as xr # cargar libreria 
nc = xr.open_dataset('grid_0.4min.nc')#abrir archivo .nc
#dsloc = ds.sel(lon=22.9,lat=40.62,method='nearest')# ingresar latitud y longitud, extrae serie en ese punto
#print(dsloc['eta'])
```

```python
#ver variables que trae el archivo .nc
for i in nc.variables.keys():
    print(i)
```

```python
#guardar variables de interes
grid_lon=nc.variables['grid_lon'][:].data
grid_lat=nc.variables['grid_lat'][:].data
lon=nc.variables['lon'][:].data
lat=nc.variables['lat'][:].data
original_bathy= nc.variables['original_bathy'][:].data
deformed_bathy= nc.variables['deformed_bathy'][:].data
max_height=nc.variables['max_height'][:].data
```

```python
#variable a plotear
bathy=deformed_bathy
```

```python
#Definicion de colormap

#maximo valor batimetria, para centrar el colormap respecto al cero
c_max=np.amax(abs(bathy))

#armar colormap para agua y tierra
c_agua=plt.cm.Blues_r(np.linspace(0,0.5,128))
c_terr=plt.cm.Oranges_r(np.linspace(0.25,0.75,128))
cols = np.vstack((c_agua, c_terr))
cmap=cm.colors.LinearSegmentedColormap.from_list("q", cols)
```

```python
len(bathy)
```

```python
#Ploteo

%matplotlib notebook

f,a = plt.subplots()
cp=a.contourf(lon,lat,-bathy,10,cmap=cmap,vmin=-c_max, vmax=c_max)
f.colorbar(cp)

a.contour(lon,lat,-bathy,0,colors = 'k',linewidths=0.4)
a.axis('scaled')

pos = [] #aqui va guardando las ubicaciones de los click
def onclick(event):
    pos.append([event.xdata,event.ydata])
    ppx, ppy = zip(*pos) #coordenada x e y de los click
    plt.scatter(ppx,ppy,s=2,c='r',marker='x'); #inform matplotlib of the new data
    plt.draw() #redraw
f.canvas.mpl_connect('button_press_event', onclick)
f.show()
```

```python
pos
```

## marcar los puntos de análisis y luego ejecutar estas celdas.

```python
#%matplotlib inline
series = {}
cont = 0
for p in pos : 
    cont +=1
    ncloc = nc.sel(lon= p[0],lat=p[1],method='nearest')# ingresar latitud y longitud, extrae serie en ese punto
    series['serie'+str(cont)]= ncloc['eta']
    plt.figure(figsize=(8,4), dpi=80)
    ncloc['eta'].plot() #grafico
plt.show()
```

#### Se guardaron las series de los puntos seleccionados en el diccionario series con la llave 'seriei' (serie1,serie2)
