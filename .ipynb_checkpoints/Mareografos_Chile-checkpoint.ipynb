```python
#from netCDF4 import Dataset
import matplotlib.cm as cm
import matplotlib.pylab as plt
import numpy as np
import xarray as xr # cargar libreria 
from netCDF4 import Dataset

```

```python
#Abrimos mapa
data=xr.open_dataset(r"C:\Users\ASUS\Desktop\Memoria\Jupyter\bat_l0_gebco_30sa_chilecont_v2.grd")#abrir mapa

#Abrir boyas virtuales
ts_txt=open(r"C:\Users\ASUS\OneDrive - Universidad Técnica Federico Santa María\Memoria\Mareografos\LATLON_TIDE_GAUGES_corregido.txt")
ts=np.loadtxt(ts_txt,dtype=str,delimiter='\t')

#ubicación de la carpeta de guardado
d_save = r"Z:\_ MEMORIA\HySea - Modelos\test_grecia"
```

```python
for i in data.variables.keys():
    print(i)
```

```python
#Variables ts

code=ts[:,0]
lonts=ts[:,1].astype(float)
latts=ts[:,2].astype(float)
```

```python
%matplotlib notebook

bathy=data.variables['z'][:].data
lon=data.variables['x'][:].data
lat=data.variables['y'][:].data
Lon,Lat=np.meshgrid(lon,lat)

c_max=np.amax(abs(bathy))
c_agua=plt.cm.Blues_r(np.linspace(0,0.5,128))
c_terr=plt.cm.Oranges_r(np.linspace(0.25,0.75,128))
cols = np.vstack((c_agua, c_terr))
cmap=cm.colors.LinearSegmentedColormap.from_list("q", cols)

f,a = plt.subplots()
cp=a.contourf(Lon,Lat,bathy,20,cmap=cmap,vmin=-c_max,vmax=c_max)
f.colorbar(cp)

a.contour(Lon,Lat,bathy,0,colors = 'k',linewidths=0.4)
a.plot(lonts,latts,'o',color='black')
a.axis('scaled')

pos = [] #aqui va guardando las ubicaciones de los click
def onclick(event):
    pos.append([event.xdata,event.ydata])
    ppx, ppy = zip(*pos) #coordenada x e y de los click
    plt.scatter(ppx,ppy,s=2,c='r',marker='x'); #inform matplotlib of the new data
    plt.draw() #redraw
f.canvas.mpl_connect('button_press_event', onclick)
f.show()
```
