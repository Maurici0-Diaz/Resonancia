```python
import glob
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from utide import solve,reconstruct
from segmentos import segmentos
from matplotlib.dates import date2num
import os
```

```python
direccion_csv=r'Z:\00 Memoria\Tareas\tarea02\data_ioc' #direccion de carpeta con series
save_folder=r'Z:\00 Memoria\Tareas\tarea02\background_spectra' #direccion donde se guardara la serie completa

code='juan' #codigo de la ubicacion que se quiera generar la serie de tiempo
lat_mareografo=-33.636017
```

```python
#juntar series en variables unicas tiempo, prs, rad

lineas=[]

for direccion_completa in glob.glob(os.path.join(direccion_csv,code+'*')):
    print(direccion_completa)

    with open(direccion_completa, 'r') as reader:
        for line in reader.readlines():
            line = line.replace('\n','')
            line = line.replace('\t',' ')
            line=line.rsplit(sep=' ', maxsplit=2)
            lineas.append(line)
lineas=np.array(lineas)

tiempo, prs, rad =lineas[:,0],lineas[:,1],lineas[:,2]

tiempo=tiempo.astype(np.datetime64)
prs = prs.astype(np.float)
rad = rad.astype(np.float)
```

```python
#aplicar utide a serie prs

time=date2num(tiempo)
u=prs

coef = solve(time, u,v=None,lat=lat_mareografo)
tide = reconstruct(time, coef)
```

```python
#guardar serie completa en .txt
#guarda columnas ["tiempo" "u" "tide.h" "u-tide.h"]

t0=np.datetime_as_string(tiempo[0], unit='D')
tf=np.datetime_as_string(tiempo[-1], unit='D')
save_name='utide_'+code+'_'+t0+'to'+tf+'.txt' #utide_coqu_2020-07-02to2020-08-01.txt


np.savetxt(os.path.join(save_folder,save_name), np.transpose((tiempo,u,tide.h,u-tide.h)),fmt='%s')
```

```python
#grafica algun segmento de la serie completa

fecha_inicio='2020-03-25'
periodo=24*60*60

i_t0=np.where(tiempo==np.datetime64(fecha_inicio))[0][0]
i_t1=np.where(tiempo==(np.datetime64(fecha_inicio)+np.timedelta64(periodo,'s')))[0][0]

fig1, (ax1, ax2, ax3) = plt.subplots(nrows=3, figsize=(15, 7),sharey=True,sharex=True)

ax1.plot(tiempo[i_t0:i_t1], u[i_t0:i_t1],label='Mareografo')
ax1.legend(numpoints=1)
ax1.grid()

ax2.plot(tiempo[i_t0:i_t1], tide.h[i_t0:i_t1],label='UTide')
ax2.legend(numpoints=1)
ax2.grid()

ax3.plot(tiempo[i_t0:i_t1],(u-tide.h)[i_t0:i_t1],label='Residuo')
ax3.legend(numpoints=1)
ax3.grid()
```
